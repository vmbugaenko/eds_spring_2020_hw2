```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
n_s = np.arange(80, 160, step=5)
passed_num = []
s_corrs = []
f_corrs = []
for n in n_s:
    rus = np.random.normal(60, 10, 10000)
    math = np.random.normal(60, 10, 10000)
    total = rus+math 
    passed = total[total >= n]
    failed = total[total < n]
    passed_num.append(len(passed))
    s_corrs.append(np.corrcoef(rus[total >= n], math[total >= n])[1, 0])
    f_corrs.append(np.corrcoef(rus[total < n], math[total < n])[1, 0])
```

```python
plt.plot(n_s, passed_num)
```

```python
plt.plot(n_s, s_corrs)
plt.title('корреляция между русским и матешей у прошедших')
plt.ylabel('корреляция')
plt.xlabel('проходной балл')
plt.show()
```

```python
plt.plot(n_s, f_corrs)
plt.title('корреляция между русским и матешей у не прошедших')
plt.ylabel('корреляция')
plt.xlabel('проходной балл')
plt.show()
```
